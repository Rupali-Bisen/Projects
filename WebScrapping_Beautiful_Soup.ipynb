### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
#Checking for length of each attribute
len(Movie_Name), len(IMDB_Ratings), len(Year_of_release)
```

```python
# Creating a dataframe

import pandas as pd
imdb_top100_movies=pd.DataFrame({})
imdb_top100_movies['Movie_Name']=Movie_Name
imdb_top100_movies['IMDB_Ratings']=IMDB_Ratings
imdb_top100_movies['Year_of_release']=Year_of_release
```

#### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
page_weather_forecast=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YLYwvfkzbIU')
page_weather_forecast
```

```python
s8=BeautifulSoup(page_weather_forecast.content, "html.parser")
```

```python
#Scrapping Period
period=s8.find_all('p',class_="period-name")


Period=[]
for i in period:
    Period.append(i.text)

## Scrapping short description
s_d=s8.find_all('p',class_="short-desc")

Short_Desc=[]
for i in s_d:
    Short_Desc.append(i.text)
Short_Desc

### Scrapping temperature
temp=s8.find_all('p',class_="temp temp-high")

Temperature=[]
for i in temp:
    Temperature.append(i.text)

## Scrapping description
descr=s8.find_all('div',class_="col-sm-10 forecast-text")

Description=[]
for i in descr:
    Description.append(i.text)
Description

len(Period),len(Short_Desc),len(Temperature),len(Description)
```

```python
#Framing Dataframe
Weather_forecast=pd.DataFrame()
Weather_forecast['Period'] = Period[:4]
Weather_forecast['Short_Desc'] = Short_Desc[:4]
Weather_forecast['Temperature'] = Temperature[:4]
Weather_forecast['Description'] = Description[:4]

Weather_forecast
```
